# Определение стоимости автомобилей для «Не бит, не крашен»

<b>Задача:</b> Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. В распоряжении исторические данные: технические характеристики, комплектации и цены автомобилей. Нужно построить модель для определения стоимости.

<b>Целевой признак:</b>
- Price — цена (евро)

## Предобработка данных

Признак NumberOfPictures был добавлен по ошибке, либо при выгрузке данных произошел сбой. Стоит удалить полностью нулевой столбец и сообщить заказчику.

```python
data = data.drop('NumberOfPictures', axis=1)
```

Из всех временных рядов может оказаться полезным только время выставления объявления. Время скачивания анкеты и дата последнего сенса пользователя бесполезны.

```python
def year_month_day (data, column):
    
    data['year'] = data[column].dt.year
    data['month'] = data[column].dt.month
    data['day'] = data[column].dt.day
    
year_month_day (data, 'DateCreated')
```

Для более точной обработки выбросов будет использовано ручное кодирование, а не правило трех сигм.

- Признак RegistrationYear до 1990 года имеет много пропусков и неточностей в данных, как и года выше 2015. 95% автомобилей, зарегистрированных между 2015-2018 годов имеют по несколько пропусков в данных. Компанией разрешено отбросить некорректно заполненные данные после 2015 года.

- Признак Power имеет больше 40.000 нулевых значений, из-за чего простое их удаление недопустимо. Заполним все нулевые значения пропусками, чтобы потом предсказать ближайшими соседями. Мощность свыше 450 считается выбросом.

- Признак Price целевой, но имеет некорректные данные. Больше 10.000 строк имеют нулевое значение, что вызвано желанием пользователя торговаться за его автомобиль. Поскольку эти данные не несут никакой пользы модели машинного обучения, а заполнять их ближайшими соседями некорректно, то остается лишь удалить данные стоимостью меньше 50.

```python
# удалим данные, где год меньше 1980 и больше 2015
data = data[(data['RegistrationYear'] > 1980) & (data['RegistrationYear'] < 2016)]

# заполним пропусками данные, где мощность меньше 45 и удалим, где мощность больше 450
data = data[data['Power'] < 450]; data.loc[data['Power'] < 45, 'Power'] = np.nan

# удалим данные, где цена меньше 50
data = data[data['Price'] > 50]

data = data.reset_index(drop=True)
print('Было удалено:', 354369 - data.shape[0], 'строк')
```

Перед вами алгоритм заполнения пропусков ближайшими соседями с добавлением в каждой итерации заполненных признаков. Другими словами, признак NotRepaired обучался на изначальных признаках без пропусков + признаки заполненные ближайшими соседями. Все признаки расположены в порядке прибавления сложности, чтобы минимизировать ошибки предсказания. Из алгоритма исключен признак Price, из-за его случайности, что влияло негативно на точные предсказания.

Алгоритм считается около 20 минут, из-за чего в конце происходит сохранение результатов  в csv файл. При перезапуске можно просто прочитать файл и не запускать алгоритм.

```python
%%time
# алгоритм заполнения пропусков ближайшими соседями
column = ['year', 'month', 'day', 'RegistrationYear', 'RegistrationMonth', 'Kilometer', 'PostalCode']

k_data = data[column]
k_data = pd.DataFrame(MinMaxScaler(feature_range=(0, 1)).fit_transform(k_data), columns=column)
k_data['Brand'] = pd.DataFrame(
    MinMaxScaler(feature_range=(0, 1)).fit_transform(pd.DataFrame(pd.factorize(data['Brand'])[0])), columns=['Brand'])

def k_distance (column):
    
    if data[column].dtypes == 'float64' or data[column].dtypes == 'int64':
        return k_data.join(pd.DataFrame(
            MinMaxScaler(feature_range=(0, 1)).fit_transform(pd.DataFrame(data[column])), columns=[column]))
        
    elif data[column].dtypes == 'object':
        return k_data.join(pd.DataFrame(
            MinMaxScaler(feature_range=(0, 1)).fit_transform(pd.DataFrame(pd.factorize(data[column])[0])), columns=[column]))
    
    
    
def k_neighbors (column):
    
    k_features = k_data[data[column].isna() == False]
    k_target = data[data[column].isna() == False][column]
    k_nan = k_data[data[column].isna() == True]
    
    model = KNeighborsClassifier(5, weights='distance', n_jobs=-1)
    model.fit(k_features, k_target)
    predictions = model.predict(k_nan)
    
    data.loc[data[column].isna() == True, column] = predictions
    
    return k_distance (column)

for i in ['Gearbox', 'FuelType', 'Model', 'VehicleType', 'Power', 'NotRepaired']:
    k_data = k_neighbors (i)

print('После заполнения осталось', data.isna().sum().sum(), 'пропусков')    

# чтобы не считать все заново, сохраним выданные ответы в файл 
data.to_csv('k_data.csv')

winsound.MessageBeep() 
```

## Обучение моделей

```python
features = data.drop('Price', axis=1)
target = data['Price']
```

```python
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, train_size = 0.8, random_state = 20)
```

```python
# разобьем столбцы на категориальные и числовые
num = []
cat = []

def data_type (data):
    
    for i in range(data.shape[1]):
        
        column = data.columns[i]
        
        if data[column].dtypes == 'float64' or data[column].dtypes == 'int64':
            num.append(data.columns[i])
            
        elif data[column].dtypes == 'object':
            cat.append(data.columns[i])
        
data_type (features)
```

При помощи grid_search были найдены лучшие из представленных параметров. Поиск гиперпараметров занял свыше двух часов. Лучшие параметры представлены в кросс валидации.

```python
#model = CatBoostRegressor(cat_features=cat, loss_function='RMSE', verbose=False)
#
#grid = {'learning_rate': [0.15, 0.1, 0.05],
#        'depth': [5, 10, 15],
#        'l2_leaf_reg': [3, 6, 9],
#        'iterations': [100, 200, 300]}
#
#model = model.grid_search(grid, X=features_train, y=target_train, cv=3, verbose=False, plot=True);
```

```python
pool = Pool(data=features, label=target, cat_features=cat)

params = {'iterations': 750,
          'depth': 10,
          'learning_rate': 0.15,
          'l2_leaf_reg': 3,
          'loss_function': 'RMSE',
          'verbose': False}

scores = cv(pool, params,  fold_count=2, shuffle=True, logging_level='Silent', plot='True')
```

```python
model = CatBoostRegressor(iterations=750, depth=10, learning_rate=0.15, l2_leaf_reg=3, 
                          cat_features=cat, loss_function='RMSE', verbose=False)

model.fit(features_train, target_train, cat_features=cat, plot=True, silent=True);
```

```python
predictions = model.predict(features_valid)
print('RMSE:', mean_squared_error(target_valid, predictions) ** 0.5)
```
